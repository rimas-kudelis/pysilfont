## Introduction

Jupyter interacts with python using an `ipython` module which does all the heavy lifting of producing nice looking output. [pysilfont](https://github.com/silnrsi/pysilfont) has an ipython module that adds some useful functions for use in a Jupyter context. After installing pysilfont and its dependencies you can go through the following examples:

```python
from silfont.ipython import *
```

Let's load up a font (adjust the example path after having downloaded the the Awami font from https://software.sil.org/awami/). The result is a fonttools `TTFont` object that can be useful for all kinds of things.

```python
f = loadFont('/home/nico/.fonts/AwamiNastaliq-Regular.ttf')
```

Yes, the plural here implies you can ask for a list of glyphs, or just specify one glyph.

```python
b = 'Beh'
displayGlyphs(f, ['abs'+b, 'abs'+b+'Ini', 'abs'+b+'Med', 'abs'+b+'Fin'])
```

More usefully, we can display correctly shaped text, that has been processed using harfbuzz.

```python
atext = u"ارک آسُودگی آوارگی اَوگُوستُس اَیساگُمراہ آتاکہ"
displayText(f, atext)
```

```python
atext
```

```python
displayRaw(atext)
```

Three different ways of looking at the same string.


# Text Processing
Let's load a font like Ezra from https://software.sil.org/ezra/ and check a common word that often has had problems rendering in the past: Jerusalem.

```python
f1 = loadFont('/home/nico/.fonts/SILEOTSR.ttf')
t = u'\u05d9\u05b0\u05e8\u05d5\u05bc\u05e9\u05c7\u05c1\u05dc\u05b7\u034f\u05b4\u0597\u05dd'
displayText(f1, t, size=40)
```
